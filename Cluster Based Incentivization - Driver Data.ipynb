# Problem Statement

Lithion power is the largest producer of eletric vehicle(e-vehicle) batteries.
They provid batteries on rent to e-vehicle drivers. Drivers rent a battery typically for a day and thereafter replacing it with a charged battery from the company.
Lithion power has a variable pricing model based on the driver's driving history. Battery life depends on factors such as over speeding, distance driver per day, etc.
You are supposed to create a cluster model where drivers can be grouped together based on the driving data and group the data points so that drivers will be incentivized based on the cluster.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (12,6)
```

```python
df = pd.read_csv('Datasets/driver-data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 2)
df_analyze = df.drop('id', axis = 1)
```

```python
kmeans.fit(df_analyze)
```

```python
kmeans.cluster_centers_
```

```python
print(kmeans.labels_)
print(len(kmeans.labels_))
```

```python
print(type(kmeans.labels_))
unique, counts = np.unique(kmeans.labels_, return_counts = True)
print(dict(zip(unique, counts)))
```

```python
df_analyze['cluster'] = kmeans.labels_
sns.set_style('whitegrid')
sns.lmplot(x = 'mean_dist_day', y ='mean_over_speed_perc', data = df_analyze
          ,hue='cluster', palette='coolwarm', height=6, aspect=1, fit_reg=False)

```

```python
kmeans_4 = KMeans(n_clusters = 4)
kmeans_4.fit(df.drop('id', axis = 1))
kmeans_4.fit(df.drop('id', axis = 1))
print(kmeans_4.cluster_centers_)
unique, counts = np.unique(kmeans_4.labels_, return_counts = True)

kmeans_4.cluster_centers_
print(dict(zip(unique, counts)))
```

```python
df_analyze['cluster'] = kmeans_4.labels_
sns.set_style('whitegrid')
sns.lmplot(x = 'mean_dist_day', y ='mean_over_speed_perc', data = df_analyze
          ,hue='cluster', palette='coolwarm', height=6, aspect=1, fit_reg=False)
```
